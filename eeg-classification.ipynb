<h1> EEG Dataset

EEG - general information
- records electrical activity of the brain
- electrical activity is generated by firing of neurons in the brain
- electrical signals
- represented as wave forms
- diagnosing neurological condtions


About dataset

- EEG recordings of subjects before and during performance of arithmetic tasks (subtract 2 digit number from 4 digit number)
- duration - 60 seconds
- 36 csv files for each subject having 19 channels
- 19 selected channels - columns - Fp1, Fp2, F3, F4, F7, F8, T3, T4, C3, C4, T5, T6, P3, P4, O1, O2, Fz, Cz, and Pz.
- The channels listed in the dataset represent specific locations on the scalp where the EEG electrodes were placed to record electrical activity. Each channel captures the electrical signals generated by the underlying brain regions.
- rows represent different samples
- EEG recordings during the last three minutes of the rest period and the first minute of the mental arithmetic performance.
- 180 for resting state and 60s for mental counting

<h3> Exploratory Data Analysis on the dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

-Channels Fz and F4 are highly correlated. <br>
-Channels P3 and Pz are also positively correlated

```python
num_rows = 5
num_cols = 4

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 15))
fig.suptitle('EEG Signals for All Channels')

channels = df.columns
for i, channel in enumerate(channels):
    row = i // num_cols
    col = i % num_cols
    axes[row, col].plot(df[channel])
    axes[row, col].set_title(channel)
    axes[row, col].set_xlabel('Time')
    axes[row, col].set_ylabel('EEG Amplitude')

fig.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

-with time the amplitude is seen to increase, i.e when the mathematical sum was given
